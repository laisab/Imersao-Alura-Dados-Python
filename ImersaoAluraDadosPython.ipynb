```python
# Dictionary for translating column names
translation_dict = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Rename the columns
df.rename(columns=translation_dict, inplace=True)

# Display the translated columns
display(df.columns)
```

```python
# Nível de experiência mais frequente na base de dados
df["senioridade"].value_counts()
```

```python
# Tamanho de empresa mais frequente na base de dados
df["tamanho_empresa"].value_counts()
```

```python
senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(senioridade)
df['senioridade'].value_counts()
```

```python
contrato = {
    'FT': 'Tempo Integral',
    'PT': 'Tempo Parcial',
    'CT': 'Contrato',
    'FL': 'Freelancer'
}

df['contrato'] = df['contrato'].replace(contrato)
df['contrato'].value_counts()
```

```python
remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

df['remoto'] = df['remoto'].replace(remoto)
df['remoto'].value_counts()
```

```python
tamanho_empresa = {
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

***
**Tratamento de Dados Nulos**
***
*   Preenchimento de valores

1) Preencher com a média ou a mediana dos dados

2) Repetir valores

3) Preencher com valores fixos

*  Excluir valores


**fillna**: preenchimento de valores nulos

```python
# Criação de um novo DataFrame para usar como exemplo
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

# Preenchimento de dados nulos com a média dos salários
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

# Preenchimento de dados nulos com a mediana dos salários
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

**ffill**: preenche com o valor anterior

```python
# Criação de um novo DataFrame para usar como exemplo
df_temperaturas = pd.DataFrame({
    'dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

# Preenchimento com valores repetidos
df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()
df_temperaturas
```

**bfill**: preenche com o valor posterior

```python
df_temperaturas = pd.DataFrame({
    'dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

# Preenchimento com valores repetidos
df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()
df_temperaturas
```

```python
# Criação de um novo DataFrame para usar como exemplo
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Belém']
})

# Preenchimento com valores fixos
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
df_cidades
```

```python
# Plotar gráfico com a própria biblioteca Pandas
df_limpo['senioridade'].value_counts().plot(kind='bar', title='Distribuição de Senioridade')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Gráficos de barra**

```python
sns.barplot(data=df_limpo, x='senioridade', y='usd')
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df_limpo, x='senioridade', y='usd')
plt.title('Salário médio por senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salário médio anual (USD)')
plt.show() # Mostra somente o gráfico
```

```python
# Cálculo da média do salário de cada categoria em ordem decrescente
df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False)
```

```python
ordem = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False).index
ordem
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df_limpo, x='senioridade', y='usd', order=ordem)
plt.title('Salário médio por senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salário médio anual (USD)')
plt.show()
```

**Histograma**

**bins**: intervalos gerados entre as barras do histograma.

**kde**: linha adicionada ao gráfico para ver a distribuição dos valores.

```python
plt.figure(figsize=(10, 5)) # Largura, altura
sns.histplot(df_limpo['usd'], bins=50, kde=True)
plt.title('Distribuição dos salários anuais')
plt.xlabel('Salário (USD)')
plt.ylabel('Frequência')
plt.show()
```

**Boxplot**

Os "pontinhos" do boxplot são outliers.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_limpo['usd'])
plt.title('Boxplot Salário')
plt.xlabel('Salário (USD)')
plt.show()
```

```python
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']

plt.figure(figsize=(8, 5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade)
plt.title('Boxplot da distribuição por senioridade')
plt.xlabel('Salário (USD)')
# plt.ylabel('USD')
plt.show()
```

```python
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']

plt.figure(figsize=(8, 5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade, palette='Set2', hue='senioridade')
plt.title('Boxplot da distribuição por senioridade')
plt.xlabel('Salário (USD)')
# plt.ylabel('USD')
plt.show()
```

**Gráficos interativos**

```python
import plotly.express as px
```

```python
senioridade_media_salario = df_limpo.groupby('senioridade')['usd'].mean().reset_index()
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']

fig = px.bar(senioridade_media_salario,
             x='senioridade',
             y='usd',
             title='Média salarial por senioridade',
             labels={'senioridade': 'Senioridade', 'usd': 'Média salarial anual (USD)'},
             category_orders={'senioridade': ordem_senioridade})

fig.show()
```

**Gráfico de pizza**

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(
    remoto_contagem,
    names='tipo_trabalho',
    values='quantidade',
    title='Proporção dos tipos de trabalho'
)

fig.show()
```
