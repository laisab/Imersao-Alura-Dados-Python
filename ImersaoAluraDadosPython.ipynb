````python
tamanho_empresa = {
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['tamanho_empresa'].value_counts()
````

***
**Tratamento de Dados Nulos**
***
*   Preenchimento de valores

1) Preencher com a média ou a mediana dos dados

2) Repetir valores

3) Preencher com valores fixos

*  Excluir valores


**fillna**: preenchimento de valores nulos

````python
# Criação de um novo DataFrame para usar como exemplo
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

# Preenchimento de dados nulos com a média dos salários
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

# Preenchimento de dados nulos com a mediana dos salários
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
````

**ffill**: preenche com o valor anterior

````python
# Criação de um novo DataFrame para usar como exemplo
df_temperaturas = pd.DataFrame({
    'dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

# Preenchimento com valores repetidos
df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()
df_temperaturas
````

**bfill**: preenche com o valor posterior

````python
df_temperaturas = pd.DataFrame({
    'dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

# Preenchimento com valores repetidos
df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()
df_temperaturas
````

````python
# Criação de um novo DataFrame para usar como exemplo
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Belém']
})

# Preenchimento com valores fixos
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
df_cidades
````

**Pycountry**: biblioteca usada para acessar dados de países, como nomes, códigos e moedas, de forma padronizada. É útil quando se trabalha com informações geográficas.

````python
import pycountry
````

````python
# Função para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
      return pycountry.countries.get(alpha_2=code).alpha_3
    except:
      return None
````

````python
# Criação de uma nova coluna com código ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)
df_limpo.head()
````

````python
# Cálculo da média salarial por país
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

df_ds.head()
````

````python
fig = px.choropleth(
    media_ds_pais,
    locations='residencia_iso3',
    color='usd',
    color_continuous_scale='rdylgn',
    title='Salário médio de Cientista de Dados por país',
    labels={'usd': 'Salário médio (USD)', 'residencia_iso3': 'País'}
)

fig.show()
````

````python
# Cálculo da média salarial por país
df_ds = df_limpo[df_limpo['cargo'] == 'Solutions Engineer']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

df_ds.head()
````

````python
fig = px.choropleth(
    media_ds_pais,
    locations='residencia_iso3',
    color='usd',
    color_continuous_scale='rdylgn',
    title='Salário médio de Engenheiro de Soluções por país',
    labels={'usd': 'Salário médio (USD)', 'residencia_iso3': 'País'}
)

fig.show()
````

````python
# Cálculo da média salarial por país
df_ds = df_limpo[df_limpo['cargo'] == 'Principal Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

df_ds.head()
````

````python
fig = px.choropleth(
    media_ds_pais,
    locations='residencia_iso3',
    color='usd',
    color_continuous_scale='rdylgn',
    title='Salário médio de Cientista de Dados Principal por país',
    labels={'usd': 'Salário médio (USD)', 'residencia_iso3': 'País'}
)

fig.show()
````
